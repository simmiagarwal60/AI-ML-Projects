```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=42)
```

```python
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
```

```python
df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
inertia = []
k_range = range(1,10)
```

```python
for k in k_range:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X_scaled)
  inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(k_range, inertia, marker='o')
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(x=df['Feature1'], y=df['Feature2'], hue = df['cluster'], palette='viridis')
```

```python
from sklearn.datasets import make_moons
```

```python
X , y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
df = pd.DataFrame(X, columns = ['Feature1', 'Feature2'])
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)
```

```python
df['kmeans_cluster'] = cluster_labels
```

```python
sns.scatterplot(x=df['Feature1'], y=df['Feature2'], hue= df['kmeans_cluster'], palette='viridis')
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=df['Feature1'], y=df['Feature2'], hue=dbscan_labels, palette='viridis')
```
